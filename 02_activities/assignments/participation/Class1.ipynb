```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)
x = np.arange(50)  #making a range of 50 using numpy so 0 to 49
y = np.random.randint(0,100, 50)   # random integers where min is 0, max is 100 and we want 50 of those values

fig, ax = plt.subplots(figsize=(5,3))       #5 and 3 are figure sizes, default is in inches
#ax.scatter(x,y)   #this is a scatter plot
#ax.bar(x,y)
#ax.plot(x,y)
ax.hist(y) #don't need to use x for histogram)
```

```python
font1 = {'family':'sans-serif', 'color':'blue','size':20}    #this is our dictionary to use it at the bottom after
font2 = {'family':'monospace', 'color':'green', 'size': 14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict=font1, loc = 'left')  #loc will be the location, default is center
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
plt.show()
```

```python
fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y,
            marker ='*',   #other options are +, D,can google to see the other options)
            color = 'indigo')
```

```python
fig, ax =plt.subplots(figsize=(5,3))
ax.plot(x,y,
            marker ='*',
            color = 'indigo',  #also can use #7425b9
            linestyle = '--',      #when we are putting a linestyle and width, need to change the graph to a plot instead of a scatter
            linewidth = 2,
            markersize = 12,
            markeredgecolor = '#fa9359', #this is the color of the star out which is a yellow. google hexcodepicker if you want to pull up what the codes mean
            markerfacecolor = '#000000')   #this is black

#ax.grid(axis ='y') # can just put ax.grid() if we want to show both x and y.

ax.grid(axis= 'y', color = 'pink',
                    linestyle = '--',
                    linewidth = 5)
```
