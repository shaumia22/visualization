```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.set_style('whitegrid') #can also put darkgrid or ticks or dark and white

tipgraph = sns.lineplot(data=tips,
             x ='total_bill',
             y='tip')

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip amount ($)')
```

```python
fig = plt.subplots(figsize=(10, 3))

tipgraph = sns.lineplot(data=tips,
                        x='total_bill',
                        y='tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth = 3,
                        marker = 'o',
                        markerfacecolor = 'indigo')

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip amount ($)')
```

```python
tipgraph = sns.scatterplot(data=tips, x='total_bill',
                           y='tip',
                           style = 'time', #style is for shape
                           hue ='day', #this makes it change the color at the different times of the day
                           palette = ['purple','hotpink','deepskyblue','yellowgreen'])

tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
sns.pairplot(data = tips, hue = 'day') #this won't use the categorical data as it can't make scatter plots. This is using the default colors and pairplot default plots. This compares variable vs variable.
```

```python
daysplot = sns.relplot(data=tips, #used to compare specific levels within variables. This is identifying the dataset to use
                       x="total_bill", #defining x variable
                       y="tip", #defining y variable
                       hue="sex", #defining what to separate the data by - i.e. diff sex will have a unique color
                       col="day", #each one is showing per day
                       kind="scatter", #plot type
                       col_wrap=2) #show only 2 columns per row.
```

```python
import plotly.graph_objects as go

x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
graph = go.Figure()

graph.add_trace(go.Bar(x=x1, y=y1))

graph.update_layout(
    title="Pirate Scores",
    xaxis_title = "Pirates",
    yaxis_title = "Score"
)
```

```python
graph = go.Figure()

graph.add_trace(go.Scatter(x=x1,
                           y=y1,
                           mode = 'markers',
                           marker = dict(
                               size = 15,
                               color = 'hotpink',
                               opacity = 1,  #shows transparency, 1 is max
                               line = dict(width =5, color= 'purple')
                               )))

graph.update_layout(title = "Interactive Pirate Plot",
                    xaxis_title = 'Pirates',
                    yaxis_title = 'Scores',
                    width = 500,
                    height = 500)
```

```python
#!pip install wordcloud
from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv", on_bad_lines = 'skip')

df
```

```python
text = " ".join (each for each in df.quote) #this will join all the text under quote column with a space

wordcloud = WordCloud(background_color= 'white',
                      colormap = 'inferno').generate(text)

fig, ax = plt.subplots(figsize =(7,3))
ax.imshow(wordcloud)

ax.axis('off') #this is to make sure we dont get any grids
```

```python
#!pip install matplotlib_venn

from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(['apple', 'banana', 'watermelon'])
B = set (['pumpkin', 'blueberry', 'apple', 'keylime'])

diagram = venn2([A,B],
                set_labels = ('Fruits', 'Pires'),
                set_colors = ('blue', 'red'),
                alpha =0.5)

diagram.get_label_by_id("10").set_text("\n".join(A-B)) #left. numbers refer to specific positioning
diagram.get_label_by_id("11").set_text("\n".join(A&B)) #middle
diagram.get_label_by_id("01").set_text("\n".join(B-A)) #right
```
